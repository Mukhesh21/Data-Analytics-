```python
missing_counts = bird.isnull().sum()
print("\nMissing values per attribute (column):")
print(missing_counts)
```

```python
#Filling missing Values by ffill
bird["Interrupted_Reason"] = bird["Interrupted_Reason"].fillna(method='ffill')
print("Number of Null values after filling:",bird["Interrupted_Reason"].isnull().sum())
bird["Interrupted_Reason"]
```

```python
unique_Migration_Reason = bird['Migration_Reason'].unique()
unique_Migration_Reason
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
bird['Encoded_Migration_Reason'] = encoder.fit_transform(bird['Migration_Reason'])
print("\nEncoded Data:\n",bird['Encoded_Migration_Reason'])
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')


encoded_data = encoder.fit_transform(bird[['Habitat']])
print("\nOne-Hot Encoded Data:\n",encoded_data)
```
